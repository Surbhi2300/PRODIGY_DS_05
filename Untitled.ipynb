```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import folium
from folium.plugins import HeatMap
import datetime
```

```python
df = pd.read_csv(r"C:\Users\Surbhi\Downloads\US_Accidents_March.csv", nrows=1000)
print(df.shape)
```

```python
df = df[['Start_Time', 'Weather_Condition', 'Temperature(F)', 'Humidity(%)', 
         'Visibility(mi)', 'Severity', 'Start_Lat', 'Start_Lng', 'Street']]

df.dropna(subset=['Start_Time', 'Weather_Condition', 'Start_Lat', 'Start_Lng'], inplace=True)
```

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
df['Hour'] = df['Start_Time'].dt.hour
df['Day/Night'] = df['Hour'].apply(lambda x: 'Night' if x < 6 or x > 18 else 'Day')
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Hour', palette='coolwarm')
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Accidents")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_weather = df['Weather_Condition'].value_counts().nlargest(10).index
sns.countplot(data=df[df['Weather_Condition'].isin(top_weather)], 
              x='Weather_Condition', order=top_weather, palette='viridis')
plt.title("Top 10 Weather Conditions during Accidents")
plt.xticks(rotation=45)
plt.ylabel("Number of Accidents")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Day/Night', palette='Set2')
plt.title("Accidents: Day vs Night")
plt.grid(True)
plt.show()
```
